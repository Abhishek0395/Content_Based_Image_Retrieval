```python
import cv2
import numpy as np
def get(image, idx, idy):
    if idx < (len(image)) and idy < len(image[0]) and idx>=0 and idy >=0:
        return 1
    else:
        return 0
def dir(x,y,img):
	ax=int(img[x+1,y])-int(img[x,y])
	ay=int(img[x,y-1])-int(img[x,y])
    
	if ax>=0 and ay>=0:
		return 1
	elif ax<0 and ay>=0:
		return 2
	elif ax<0 and ay<0:
		return 3
	elif ax>=0 and ay<0:
		return 4
		
def tetra(x,y,img):
	out=[]
	fx=[0,-1,-1,-1,0,1,1,1]
	fy=[1,1,0,-1,-1,-1,0,1]
	val=dir(x,y,img)
	for i in range(0, 8):
		a=dir(x+fx[i],y+fy[i],img)
		#if x==2 and y==2 :
			#print(val," ",a," ",x+fx[i],"\n")
		if a==val:
			out.append(0)
		else:
			out.append(a)
	return out,val

def add_dict(dic,val1,val2,val3,direc):
    #print("val + direc = "+str(val)+"  "+ str(direc))
    d=1
    val=0
    for i in range(1,4):
        if(d!=direc):
            if(i==1):
                val=val1
            elif(i==2):
                val=val2
            elif(i==3):
                val=val3
            if (1000*d+val) not in dic.keys():
                dic[25000*d] = dic[25000*d] +1
            else:
                dic[d*1000+val] = dic[d*1000+val] + 1
        d=d+1;
            

def local_tetra_pattern(img,dic1, dic2, dic3, dic4):
    for x in range(2, len(img)-2):
        for y in range(2, len(img[0])-2):
            value, direc = tetra(x,y,img)
            val1=0
            val2=0
            val3=0
            val4=0
            for i in range(0,len(value)):
                if(value[i]==1):
                    val1+=2**(7-i)
                elif(value[i]==2):
                    val2+=2**(7-i)
                elif(value[i]==3):
                    val3+=2**(7-i)
                elif(value[i]==4):
                    val4+=2**(7-i)
            dc = dir(x, y, img)
            if dc != direc:
                print("WTF = " + str(direc) + " "+ str(dc))
            #print(value)
            #print("dir and all 4 val = "+str(direc)+" "+str(val1)+"  "+str(val2)+"  "+str(val3)+"  "+str(val4))
            
            if direc == 1:
                add_dict(dic1, val2,val3,val4,direc)
            elif direc == 2:
                add_dict(dic2, val1,val3,val4, direc)
            elif direc == 3:
                add_dict(dic3,val1,val2,val4, direc)
            elif direc == 4:
                add_dict(dic4, val1,val2,val3, direc)
    return  dic1, dic2, dic3, dic4
```
