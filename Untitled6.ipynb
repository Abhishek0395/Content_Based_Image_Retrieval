```python
import cv2
import numpy as np
def add_dict(dic,val1,val2):
    val=0
    for i in range(1,3):
        if(i==1):
            val=val1
        elif(i==2):
            val=val2
        if (1000*i+val) not in dic.keys():
            dic[25000*i] = dic[25000*i] +1
        else:
            dic[i*1000+val] = dic[i*1000+val] + 1
def lap(x,y,img):
    t=12
    a=70
    out=[]
    fx=[0,-1,-1,-1,0,1,1,1]
    fy=[1,1,0,-1,-1,-1,0,1]
    for i in range(0, 8):
        if(int(img[x,y])+t>=int(img[x+fx[i]][y+fy[i]])):
            if(int(img[x,y])-t<=int(img[x+fx[i]][y+fy[i]])):
                out.append(1)
            else:
                out.append(0)
        elif(int(img[x,y])+a<=int(img[x+fx[i]][y+fy[i]])):
            out.append(-1)
        elif(int(img[x,y])-a>=int(img[x+fx[i]][y+fy[i]])):
            out.append(-1)
        else:
             out.append(0)
    return out
def local_ap_pattern(img,dic):
    for x in range(2, len(img)-2):
        for y in range(2, len(img[0])-2):
            value = lap(x,y,img)
            val1=0
            val2=0
            for i in range(0,len(value)):
                if(value[i]==1):
                    val1+=2**(7-i)
                elif(value[i]==-1):
                    val2+=2**(7-i)
            add_dict(dic,val1,val2)
    return dic
             
```

```python
def uniform(pattern):
    pat= int(pattern)
    a=0
    b=0
    cnt=0
    for i in range(0,8):
        if( i==0 ):
            a= int(pattern/2**(7-i))
        else:
            b= int(pattern/2**(7-i))
            if(b!=a):
                cnt=cnt+1
                a=b
        pattern=pattern%2**(7-i)
    if(cnt<=2):
        return 1 #uniform hbe
    else:
        return 0
items = []
true_items = []
#items.append(0)
for i in range(0, 256):
    if uniform(i):
        items.append(1*1000+i)
        true_items.append(i)
        items.append(2*1000+i)
items.append(25000)
items.append(50000)
true_items.append(25000)
```

```python
import cv2
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
#from keras.utils import np_utils
from scipy.stats import itemfreq
import time
from functools import wraps

df = pd.read_csv(shuffled_labels)
rows = df.iterrows()

X_addrs = []
X_hist = []
Y_hist = []
#row = rows[0]
j = 0

# uncomment below this for local tetra pattern
dic = {}
def create_dic():
    for i in range(len(items)):
        dic[items[i]] = 0
start_time = time.time()      
for row in rows:
    #print(row[1][1])
    create_dic()
    print(row[1][0])
    img = cv2.imread(row[1][0],0)
    #img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    dic=local_ap_pattern(img,dic)
    new_x = []
    
    for i in dic.keys():
        new_x.append(dic[i])
    new_x = np.array(new_x)
    hist = new_x/np.sum(new_x)
    #print(hist)
    X_addrs.append(row[1][0])
    X_hist.append(hist)
    Y_hist.append(row[1][1])
    elapsed_time = time.time() - start_time
    start_time = time.time()
    print(str(j) +" "+str(row[1][0])+"  label = "+str(row[1][1]) + " done in " + str(elapsed_time))
    j = j + 1
    
```
